```python
# Number of missing values per columns of data
missing_values = df.isna().sum().sort_values(ascending=False)

print(f'Num of missing values per columns(sorted): \n{missing_values}')
```

```python
# Missing value percentage

missing_value_percentage = missing_values / len(df)
# print(f'Num of missing values per columns in percentage: \n{missing_value_percentage}')

print(f'Num of missing values per columns in percentage: \n{missing_value_percentage * 100}')
```

```python
# remove the columns that have no missing values (i.e is 0) and plot the bar graph for the remaining 
missing_value_percentage[missing_value_percentage != 0].plot(kind='barh')
```

# Exploratory Analysis and Visualization

### Columns of data we will analyze
1. City
2. Start Time
3. Start Lat, Start Lng
4. Temperature
5. Weather Condition

```python
cities = df.City.unique()
len(cities)
```

```python
city_by_accidents = df.City.value_counts()
```

```python
city_by_accidents
```

```python
print(df.State.unique())
len(df.State.unique())
```

### SO our data contains 49 uniques States....It also contains DC which is not a state.....There are two states missing form the list...which are AK (Alaska) and HI (Hawaii)

```python
city_by_accidents[:30].plot(kind='barh')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))


sns.set_style('darkgrid')

sns.distplot(city_by_accidents, ax=axes[0])
axes[0].set_title('Accidents count Distribution by Cities')

sns.kdeplot(data=city_by_accidents, fill=True, alpha=0.5, ax=axes[1])
axes[1].set_title('Accidents count Distribution by Cities')

sns.histplot(city_by_accidents, ax=axes[2])
axes[2].set_yscale('log')
axes[2].set_title('Accidents count Distribution by Cities')
```

The graph shows that most of the distribution of accidents by cities fall in less then `1000` accidents

```python
high_accident_city = city_by_accidents[city_by_accidents >= 1000]
low_accident_city = city_by_accidents[city_by_accidents < 1000]
```

```python
len(high_accident_city), len(low_accident_city)
```

```python
print(f' High Accident City(more then 1000 accidents): {len(high_accident_city) / len(cities) * 100 :.4f} %')
```

```python
# These are accident count distribution for high ( >1000) and low ( <1000) accident city

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.distplot(high_accident_city, ax=axes[0])
axes[0].set_title('High Accident City')

sns.histplot(low_accident_city, ax=axes[1], log_scale=True)
axes[1].set_title('Low Accident City(Log Scale)')
```
