```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


fig, ax = plt.subplots()



# rect = Rectangle([5, 5], width=50, height=50, facecolor='none', edgecolor='red')
# ax.add_patch(rect)
# plt.plot(5, 5, 'o')

rect = Rectangle([0, 0], width=50, height=50, facecolor='none', edgecolor='red')
ax.add_patch(rect)
plt.plot(0, 0, 'o')

ax.set_xlim(0, 100)
ax.set_ylim(0, 100)

plt.show()
```
