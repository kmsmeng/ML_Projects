```python
summary_data = pd.read_csv('first_50_summary.csv')
# summary_data
```

```python
# box_loss = [x for x in summary_data['train/box_loss']]
box_loss_train = summary_data['train/box_loss']
box_loss_val = summary_data['val/box_loss']
# box_loss
```

```python
plt.plot(range(1, len(box_loss_train)+1), box_loss_train, label='Train Box Loss')
plt.plot(range(1, len(box_loss_val)+1), box_loss_val, label='Val Box Loss')
plt.xlabel('Epoch')
plt.ylabel('Box Loss')
plt.legend()
plt.title('YOLOv8 Box Loss over Epochs')
plt.grid(True)
```

```python
map50_train = summary_data['metrics/mAP50(B)']
mAP50_95_train = summary_data['metrics/mAP50-95(B)']


# plt.plot(range(1, len(map50_train)+1), map50_train, label='mAP@50 (Train)')
plt.plot(range(1, len(map50_train)+1), map50_train, marker='o', label='mAP@50 (Train)')
plt.plot(range(1, len(mAP50_95_train)+1), mAP50_95_train, label='mAP@50:95 (Train)')



plt.xlabel('Epoch')
plt.ylabel('Mean Average Precision (mAP)')
plt.title('YOLOv8 Training mAP over Epochs')
plt.legend()
plt.grid(True)
```
